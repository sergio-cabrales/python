```
#import libraries
import pandas as pd
import numpy as np
import datetime
import pandas_datareader as web
import pandas_datareader as pdr
import matplotlib.pyplot as plt
```

```
# Get the stock quote
df = web.DataReader('USDCOP=X', data_source='yahoo',start='2021-01-01', end='2021-03-30')
df
```

```
df['Close_100'] = 100*df['Close']/df['Close'][0]
df
```

```
# Get the stock quote
df2 = web.DataReader('DX-Y.NYB', data_source='yahoo',start='2021-01-01', end='2021-03-30')
df2['Close_100'] = 100*df2['Close']/df2['Close'][0]
df2
```

```
oil = pdr.get_data_fred('DCOILBRENTEU',start='2021-01-01', end='2021-03-30')
oil.columns = ['Close']
oil['Close_100'] = 100*oil['Close']/oil['Close'][1]
oil
```

```
plt.figure(figsize=(16,8))
plt.title ('Dollar')
plt.plot(df['Close_100'], color='green', alpha = 0.6)
plt.plot(df2['Close_100'], color='red', alpha = 0.6)
plt.plot(oil['Close_100'], color='black', alpha = 0.6)
plt.xlabel('date')
plt.ylabel('Index (100)')
plt.legend(['USD/COP', 'US Dollar - Index'],loc='lower right',)
plt.show()
```
